```python
import pandas as pd
```

```python
df = pd.read_csv('Top_Anime_data.csv')
```

```python
df = df.drop(792, axis = 0)
```

```python
df = df.drop('Rank', axis = 1) # -> Dataframe dropped
```

```python
df
```

```python
score = df['Score']
pop = df['Popularity']
```

```python
score_mu, score_std = score.mean(), score.std()
pop_mu, pop_std = pop.mean(), pop.std()
```

$$Z(x;\mu, \sigma) = \frac{x - \mu}{\sigma}$$

```python
df['Score'] = (df['Score'] - score_mu)/score_std
```

```python
df['Popularity'] = (df['Popularity'] - pop_mu)/pop_std
```

```python
df['New Score'] = 2 * df['Score'] * df['Popularity'] / (df['Popularity'] + df['Score'])
```

```python
df = df.sort_values('New Score', ascending = False)
```

$$MinMax(x) = \frac{x - min}{max-min}$$
$$Max(x) = \frac{x}{max}$$

```python
df['New Score'] = (df['New Score'] - df['New Score'].min())/(df['New Score'].max() - df['New Score'].min())
```
