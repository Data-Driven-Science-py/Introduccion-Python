# SubPlots
Bsicamente es un mtodo para poder hacer una figura que tenga diferentes axes, donde nosotros podemos poner en que lugar puede ir cada una y cuantos queremos.

```python
fig, ax = plt.subplots(figsize=(10,10), dpi=100, nrows=3, ncols=3)
x=np.linspace(0,10,1000)
fig.suptitle('Funciones')
for i in range(3):
    for j in range(3):
        ax[i,j].set_ylabel(f'Funcion')
        ax[i,j].set_xlabel(f'Dominio')
        ax[i,j].plot(x,np.exp((i+j)*x))
        
plt.tight_layout()
```

```python
corr=df.groupby('age')['movement_acceleration'].mean().reset_index().values
corrdf=df.groupby('age')['movement_acceleration'].mean().reset_index()
x=corr[:,0]
y=corr[:,1]
fig = plt.figure(figsize=(5,4), dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, 'r', marker='o', markerfacecolor='k', ls='dotted')
ax.set_ylim([25,70])
ax.set_title('aos vs aceleracin media por aos')
ax.set_ylabel('aceleracion')
ax.set_xlabel('age')
ax.text(10,18,'Se puede ver como existe una correlacion entre la aceleracion y la edad', rotation='horizontal')
ax.text(10,15,f'La correlacion entre los dos es de {corrdf.iloc[:,0].corr(corrdf.iloc[:, 1])}', rotation='horizontal')
ax.bar(x,y,color='orange')
ax.annotate('Existe un jugador de 54 aos', xy=(54.,29.), xytext=(50,50), arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('El pico de aceleracion media es a los 26 aos', xy=(26,68.), xytext=(35,65), arrowprops=dict(facecolor='black', shrink=0.05))
ax.text(60,70,
       r'$\int_t^5 f(x) dx$', size=18, rotation='horizontal')
```

```python
corr=df.groupby('age')['movement_agility'].mean().reset_index().values
corrdf=df.groupby('age')['movement_agility'].mean().reset_index()
x=corr[:,0]
y=corr[:,1]
fig = plt.figure(figsize=(10,6), dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, 'b', marker='o', markerfacecolor='k', ls='-.')
ax.set_ylim([20,70])
ax.set_title('aos vs agilidad en movimiento media por aos')
ax.set_ylabel('agilidad en movimiento')
ax.set_xlabel('age')
ax.text(15,15,'Se puede ver como existe una correlacion entre la aceleracion y la edad', rotation='horizontal')
ax.text(15,12,f'La correlacion entre los dos es de {corrdf.iloc[:,0].corr(corrdf.iloc[:, 1])}', rotation='horizontal')
ax.bar(x,y,color='purple')
ax.annotate('pico de agiliadad (26 aos)', xy=(26.,66.), xytext=(40,65),arrowprops=dict(facecolor='black', shrink=0.05))
ax.text(51,66,
       r'$\int_{0}^{\pi}\frac{x^{4}\left(1-x\right)^{4}}{1+x^{2}}dx =\frac{22}{7}-\pi$', size=18, rotation='horizontal')
```

Como conclusion, como podemos ver todas las caracteristicas de los porteros son inversamente proporcionales a el control del balon y el drible, por lo que podemos concluir que los porteros tienen muy mal control de balon y drible en el juego.

Por lo visto, existe un correlacion inversa entre la altura y el balance del jugador, por lo tanto podemos concluir que mientras mas altura, menos balance:

```python
fig, ax = plt.subplots(figsize=(11,11), dpi=100)
ax.scatter(np.asarray(df['height_cm']), np.asarray(df['movement_balance']))
ax.set_title('Correlacion entre altura y balance', 
             {'fontsize': 20,
        'fontweight': 'bold',
        'color': 'k',
        'verticalalignment': 'baseline'})
ax.set_ylabel('Balance', {'fontsize': 14,
        'fontweight': 'bold',
        'color': 'k',
        'verticalalignment': 'baseline'})
ax.set_xlabel('Altura', {'fontsize': 14,
        'fontweight': 'bold',
        'color': 'k',
        'verticalalignment': 'baseline'}, labelpad=15)
ax.set_facecolor('Snow')
ax.grid(True, color='lightsteelblue', dashes=(0.1,0.1, 0.2, 0.2))
x=list(np.asarray(df['height_cm']))
y=list(np.asarray(df['movement_balance']))

m, b = np.polyfit(x, y, 1)
ax.plot(x, y, 'yo', x, m*np.asarray(df['height_cm'])+b, color='k', ls='-.')

```
